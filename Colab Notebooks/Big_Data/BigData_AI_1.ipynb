```python
import matplotlib.pyplot as plt

grade = (2.43, 3.39, 3.05, 3.81, 4.05, 4.07)
year = ('1st semester 2018', '2st semester 2018', '1st semester 2019',\
        '2st semester 2022', '1st semester 2023', '2st semester 2023')

plt.figure(figsize=(15, 5))
plt.plot(year, grade, 'co-', label='Grade'); plt.title('Grade Graph');
plt.legend(); plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 50, 0.1)

plt.figure(figsize=(50, 5))
plt.ax.plot(x, np.sin(x), np.cos(x), 'r+-')
plt.title('sin graph'); plt.xlabel('x'); plt.ylabel('sin(x)'); plt.grid(True)
plt.legend('')

plt.show()
```
