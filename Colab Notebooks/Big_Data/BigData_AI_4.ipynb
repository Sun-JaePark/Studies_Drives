```python
# Linear Model Practice

import numpy as np
from sklearn.linear_model import LinearRegression

# Data preparation
x_train = np.array([[1], [2], [3], [4]])
y_train = np.array([1, 2, 3, 4])

# Test data
x_test = np.array([[10], [12]])

# Linear Model empty object
model = LinearRegression()

# Same with train
model.fit(x_train, y_train)

# Y = aX + b (a = model.coef_, b = model.intercept_)
print(model.intercept_); print(model.coef_ ) # => Y = X

pred_y = model.predict(x_test) # Use x_test, predict Y
print(pred_y)

plt.plot(x_test, pred_y)
```

```python
# Polynomial regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np


x_train = np.array([[1], [2], [3], [4], [5]])
y_train = np.array([1, 4, 9, 16, 25])

x_test = np.array([[10], [11], [12], [13], [14]])

poly_feature = PolynomialFeatures(degree=2, include_bias=True)
x_poly_train = poly_feature.fit_transform(x_train)
print(x_poly_train)

model = LinearRegression()
model.fit(x_poly_train, y_train)

# => Y = x^2
# Y = a1*x^2, a2*x + a3 + b (a1, a2, a3 = model.coef_, b = model.intercept_)
print(model.coef_); print(model.intercept_)

x_poly_test = poly_feature.fit_transform(x_test)
pred_y = model.predict(x_poly_test)
print(pred_y)

plt.plot(x_test, pred_y)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

x_train = np.arange(-10, 11).reshape(-1, 1)
y_train = x_train**2

x_test = np.arange(-100, 101).reshape(-1, 1)


model = make_pipeline(PolynomialFeatures(degree=2, include_bias=True), LinearRegression())
model.fit(x_train, y_train)

pred_y = model.predict(x_test)
print(pred_y)
print(model.steps[1][1].coef_); print(model.steps[1][1].intercept_)

plt.plot(x_test, pred_y)
```
