```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import numpy as np

X = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
Y = np.array([[-1], [1], [1], [-1]])

n_input = 2
n_output = 1
n_layer = 10

per = Sequential()
per.add(Dense(units=n_layer, activation='tanh', input_shape=(n_input,)))
per.add(Dense(units=n_output, activation='tanh'))

per.compile(loss='mse', \
            optimizer=SGD(learning_rate=.1), metrics=['mse'])
per.fit(X, Y, epochs=500, verbose=2)

res = per.predict(X)
print(res)
```
